```python
def read_json_file(file_path):
    """
    Reads a JSON file and returns its contents as a dictionary.

    :param file_path: The path to the JSON file.
    :return: A dictionary containing the JSON data.
    """
    try:
        with open(file_path, 'r') as file:
            data = json.load(file)
        return data
    except FileNotFoundError:
        print(f"Error: The file at {file_path} was not found.")
        return None
    except json.JSONDecodeError:
        print(f"Error: The file at {file_path} could not be decoded as JSON.")
        return None
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        return None
```

```python
result_path = fr'C:\Users\FPT\Desktop\ds107\ESC499-Thesis-SAT-Trees\SATreeCraft\results\wine'
```

```python
def give_results(result_path):    
    MD_result_list = read_json_file(os.path.join(result_path,'MD_result_list.json'))
    MD_time_list = read_json_file(os.path.join(result_path,'MD_time_list.json'))
    MD_CC_result_list = read_json_file(os.path.join(result_path,'MD_CC_result_list.json'))
    MD_CC_time_list = read_json_file(os.path.join(result_path,'MD_CC_time_list.json'))
    MS_MD_result_list = read_json_file(os.path.join(result_path,'MS_MD_result_list.json'))
    MS_MD_time_list = read_json_file(os.path.join(result_path,'MS_MD_time_list.json'))
    MS_MD_CC_result_list = read_json_file(os.path.join(result_path,'MS_MD_CC_result_list.json'))
    MS_MD_CC_time_list = read_json_file(os.path.join(result_path,'MS_MD_CC_time_list.json'))
    MS_MD_result_list_depth2 = read_json_file(os.path.join(result_path,'MS_MD_result_list_depth2.json'))
    MS_MD_time_list_depth2 = read_json_file(os.path.join(result_path,'MS_MD_time_list_depth2.json'))
    MS_MD_result_list_depth3 = read_json_file(os.path.join(result_path,'MS_MD_result_list_depth3.json'))
    MS_MD_time_list_depth3 = read_json_file(os.path.join(result_path,'MS_MD_time_list_depth3.json'))
    MS_MD_result_list_depth4 = read_json_file(os.path.join(result_path,'MS_MD_result_list_depth4.json'))
    MS_MD_time_list_depth4 = read_json_file(os.path.join(result_path,'MS_MD_time_list_depth4.json'))

    feasible_runs_MD = {key: sum(v < 1800 for v in value) for key, value in MD_time_list.items()}
    feasible_runs_MS_MD = {key: sum(v < 1800 for v in value) for key, value in MS_MD_time_list.items()}

    feasible_runs_depth2:dict = {'2': sum(v < 1800 for v in value) for key, value in MS_MD_time_list_depth2.items()}
    feasible_runs_depth3:dict = {'3': sum(v < 1800 for v in value) for key, value in MS_MD_time_list_depth3.items()}
    feasible_runs_depth4:dict = {'4': sum(v < 1800 for v in value) for key, value in MS_MD_time_list_depth4.items()}
    feasible_runs_depth = {**feasible_runs_depth2, **feasible_runs_depth3, **feasible_runs_depth4}

    result_list = [MD_result_list, MD_CC_result_list, MS_MD_result_list, MS_MD_CC_result_list]
    time_list = [MD_time_list, MD_CC_time_list, MS_MD_time_list, MS_MD_CC_time_list]

    result_list_1 = [MD_result_list, MD_CC_result_list, MS_MD_result_list, MS_MD_CC_result_list]
    time_list_1 = [MD_time_list, MD_CC_time_list, MS_MD_time_list, MS_MD_CC_time_list]

    k_list = ['0.0','0.1','0.25','0.5','1.0']

    #sort out the ivalid results
    MD_result_list = {k: [i for i in MD_result_list[k] if i > 0] for k in k_list }
    MD_time_list = {k: [i for i in MD_time_list[k] if i < 1800] for k in k_list }
    MD_CC_result_list = {k: [i for i in MD_CC_result_list[k] if i > 0] for k in k_list }
    MD_CC_time_list = {k: [i for i in MD_CC_time_list[k] if i < 1800] for k in k_list }
    MS_MD_result_list = {k: [i for i in MS_MD_result_list[k] if i > 0] for k in k_list }
    MS_MD_time_list = {k: [i for i in MS_MD_time_list[k] if i < 1800] for k in k_list }
    MS_MD_CC_result_list = {k: [i for i in MS_MD_CC_result_list[k] if i > 0] for k in k_list }
    MS_MD_CC_time_list = {k: [i for i in MS_MD_CC_time_list[k] if i < 1800] for k in k_list }
    
    MS_MD_result_list_depth2 = {k: [i for i in MS_MD_result_list_depth2[k] if i > 0] for k in ['0.5']}
    MS_MD_time_list_depth2 = {k: [i for i in MS_MD_time_list_depth2[k] if i < 1800] for k in ['0.5']}
    MS_MD_result_list_depth3 = {k: [i for i in MS_MD_result_list_depth3[k] if i > 0] for k in ['0.5']}
    MS_MD_time_list_depth3 = {k: [i for i in MS_MD_time_list_depth3[k] if i < 1800] for k in ['0.5']}
    MS_MD_result_list_depth4 = {k: [i for i in MS_MD_result_list_depth4[k] if i > 0] for k in ['0.5']}
    MS_MD_time_list_depth4 = {k: [i for i in MS_MD_time_list_depth4[k] if i < 1800] for k in ['0.5']}
    
    #count the mean of the results
    MD_result_list = {k: np.mean(MD_result_list[k]) for k in k_list }
    MD_time_list = {k: np.mean(MD_time_list[k]) for k in k_list }
    MD_CC_result_list = {k: np.mean(MD_CC_result_list[k]) for k in k_list }
    MD_CC_time_list = {k: np.mean(MD_CC_time_list[k] ) for k in k_list }
    MS_MD_result_list = {k: np.mean(MS_MD_result_list[k] ) for k in k_list }
    MS_MD_time_list = {k: np.mean(MS_MD_time_list[k] ) for k in k_list }
    MS_MD_CC_result_list = {k: np.mean(MS_MD_CC_result_list[k] ) for k in k_list }
    MS_MD_CC_time_list = {k: np.mean(MS_MD_CC_time_list[k]) for k in k_list }
    MS_MD_result_list_depth2 = {'2': np.mean(MS_MD_result_list_depth2[k]) for k in ['0.5']}
    MS_MD_time_list_depth2 = {'2': np.mean(MS_MD_time_list_depth2[k]) for k in ['0.5']}
    MS_MD_result_list_depth3 = {'3': np.mean(MS_MD_result_list_depth3[k]) for k in ['0.5']}
    MS_MD_time_list_depth3 = {'3': np.mean(MS_MD_time_list_depth3[k]) for k in ['0.5']}
    MS_MD_result_list_depth4 = {'4': np.mean(MS_MD_result_list_depth4[k]) for k in ['0.5']}
    MS_MD_time_list_depth4 = {'4': np.mean(MS_MD_time_list_depth4[k]) for k in ['0.5']}

    #combine the results
    depth_result_list = {**MS_MD_result_list_depth2, **MS_MD_result_list_depth3, **MS_MD_result_list_depth4}
    depth_time_list = {**MS_MD_time_list_depth2, **MS_MD_time_list_depth3, **MS_MD_time_list_depth4}

    data_table3 = {
        'ARI': depth_result_list,
        'Feas.': feasible_runs_depth,
        'Time (s)': depth_time_list
    }

    data_table2 = {
        'ARI (MD,MS)': MS_MD_result_list,
        'ARI (MD,MS) (CC)': MS_MD_CC_result_list,
        'Feas. (MD,MS)': feasible_runs_MS_MD,
        'Time (s) (MD,MS)': MS_MD_time_list,
        'ARI (MD)': MD_result_list,
        'ARI (MD) (CC)': MD_CC_result_list,
        'Feas. (MD)': feasible_runs_MD,
        'Time (s) (MD)': MD_time_list
    }

    table2 = pd.DataFrame(data_table2)
    table3 = pd.DataFrame(data_table3)
    return table2,table3
```
